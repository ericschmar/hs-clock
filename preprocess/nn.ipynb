```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
import pull_cards
import numpy as np
import json
from ast import literal_eval
from pprint import pprint
from keras.optimizers import SGD
from keras import optimizers


b_size = 32
input_dimension = 30

decks = pull_cards.parse_decks_json("decks.json")
deck_ids = []
deck_lists = []
for i in range(len(decks)):
    #decks[i].print_deck()
    deck_ids.append(decks[i].id)
    d = literal_eval(json.loads(decks[i].deck_list))
    d.append(decks[i].game_length)
    d.append(decks[i].turns)
    deck_lists.append(d)

np_decks = np.array(deck_ids)
np_deck_lists = np.asarray(deck_lists)
#print np_deck_lists
#print np_decks
d = []
for i in range(len(np_deck_lists)):
    #print e
    e = np_deck_lists[i]
    d.append([])
    for j in range(len(e)):
        #print e[j]
        if type(e[j]) is float: continue
        if len(e[j]) == 1: continue
        if e[j][1] == 2:
            d[i].append(e[j][0])
            d[i].append(e[j][0])
        else:
            d[i].append(e[j][0])
       
    #print d[i]

print np.array(d)
one_hot_decks = keras.utils.to_categorical(d)
one_hot_labels = keras.utils.to_categorical(np_decks)
print one_hot_labels

# For a single-input model with 10 classes (categorical classification):
model = Sequential()
model.add(Dense(b_size, activation='relu', input_dim=input_dimension))
model.add(Dense(30))
model.add(Dense(30))
model.add(Dense(30))
model.add(Dense(30))
model.add(Dense(30))
model.add(Dense(173, activation='softmax'))

sgd = SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)
ada = keras.optimizers.Adagrad(lr=0.01, epsilon=1e-08, decay=0.0)
adam = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
rms = keras.optimizers.RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
model.compile(optimizer=sgd,
              loss='categorical_crossentropy',
              metrics=['accuracy'])


# Convert labels to categorical one-hot encoding
#one_hot_labels = keras.utils.to_categorical(labels, num_classes=10)

# Train the model, iterating on the data in batches of 32 samples
model.fit(np.array(d), one_hot_labels, epochs=10, batch_size=b_size, verbose=1, validation_split=0.2, shuffle=False)
```

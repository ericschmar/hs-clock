```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation, BatchNormalization
import pull_cards
import numpy as np
import json
from ast import literal_eval
from pprint import pprint
from keras.optimizers import SGD
from keras import optimizers
from keras import metrics
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

b_size = 32 #2349
# 30 cards, game length, and number of turns
input_dimension = 32

decks = pull_cards.parse_decks_json("decks.json")
deck_ids = []
deck_lists = []
for i in range(len(decks)):
    #decks[i].print_deck()
    deck_ids.append(decks[i].id)
    d = literal_eval(json.loads(decks[i].deck_list))
    d.append(decks[i].game_length)
    d.append(decks[i].turns)
    deck_lists.append(d)

np_decks = np.array(deck_ids)
np_deck_lists = np.asarray(deck_lists)
#print np_deck_lists
#print np_decks
d = []
for i in range(len(np_deck_lists)):
    #print e
    e = np_deck_lists[i]
    d.append([])
    f = []
    for j in range(len(e)):
        #print e[j]
        if type(e[j]) is float: 
            f.append(int(e[j]))
            continue
        #if len(e[j]) == 1: continue
        if e[j][1] == 2:
            f.append(e[j][0])
            f.append(e[j][0])
        else:
            f.append(e[j][0])
    d[i] = keras.utils.normalize(f)[0]

#print np.array(d)
one_hot_decks = keras.utils.to_categorical(d)
one_hot_labels = keras.utils.to_categorical(np_decks)
#print one_hot_labels
#print one_hot_decks.shape


# For a single-input model with 10 classes (categorical classification):
model = Sequential()
model.add(Dense(b_size, activation='relu', input_shape=(input_dimension, )))
model.add(Dense(30))
model.add(Dense(30))
model.add(BatchNormalization(axis=-1, momentum=0.99, epsilon=0.001, center=True, 
                                scale=True, beta_initializer='zeros', gamma_initializer='ones', 
                                moving_mean_initializer='zeros', moving_variance_initializer='ones'))
model.add(Dense(30))
model.add(Dense(30))
model.add(Dense(max(np_decks) + 1, activation='softmax'))

history = {}

sgd = SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)
ada = keras.optimizers.Adagrad(lr=0.01, epsilon=1e-08, decay=0.0)
adam = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
rms = keras.optimizers.RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
optimizers = {"sgd":sgd, "ada":ada, "adam":adam, "rms":rms}
learning_rates = [0.5, 0.1, 0.01, 0.001, 0.0001]
for k, v in optimizers.iteritems():  
    print "Using optimizer: " + k
    model.compile(optimizer=v,
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])

# Train the model, iterating on the data in batches of 32 samples
    history[k] = model.fit(np.array(d), one_hot_labels, epochs=10, batch_size=b_size, 
                           verbose=1, validation_split=0.3, shuffle=True)
    model.reset_states()
```
